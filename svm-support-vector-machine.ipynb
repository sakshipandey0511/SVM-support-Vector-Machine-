
 # SVM
* it is a supervised ml algo used for classification and regression problem
* it finds the optimal boundary (hyperplane) that separate data points of different classes with max margin
* svm is widely used in :
  1. image classification
  2. text classificattion
  3. bioinformatics
  4. handwritting recognition
     
**Hyperplane in SVM**
  a hyperplane is a decision boundary that separates  different classes
   for 2D data it is a  line , while for higherdimension it is called a    hyperplane
  
**the matmatical equation for hyperplane**
          wX+b=0
   where:
 1. w= weight
 2. x=input feature vector
 3. b=  bias
    
**Margin**
the margin is the distance between the hyperplane and the nearest data point from the both classes

**Goal of svm**
  to maximise the margine so that classification become more accurate
  
**Support vector**
  the closest data points to the hyperplane ,crucial for determining the hyperplane and margine in svm
  
**Kernel**
  a function that map data to a higher dimensional
  
**linear svm**:when data can be separeted using straight line
 
**non linear**:cannot be separated using straight line

**kernel function**:transform data to a higher dimensional space

 *steps in svm algo:*
 
    1. input training dataset
    2. convert data into feature vector
    3. choose kernal function
    4. find optimal hyperplane
    5. identify support vector
    6. maximize margin
    7. classify new data point
 

# ✅ Advantages of SVM
* Effective in high-dimensional spaces:  
Works well when the number of features is large compared to the number of samples.

* Robust to overfitting : 
Especially effective when the margin of separation between classes is large.

* Versatile with kernels : 
Can handle linear and non-linear classification using kernel tricks (e.g., polynomial, RBF).

* Strong theoretical foundation : 
Based on convex optimization, so solutions are globally optimal.

* Good generalization  :
Performs well in practice on many classification problems.

❌ Disadvantages of SVM
* Computationally expensive  :
Training time can be high for large datasets.

* Memory-intensive  :
Requires storing support vectors, which can be costly for big data.

* Choice of kernel is critical  :
Performance depends heavily on selecting the right kernel and tuning parameters.

* Not ideal for noisy data  :
Sensitive to overlapping classes and mislabeled data.

* No direct probability estimates : 
Unlike logistic regression, SVM doesn’t naturally provide probability outputs (though methods like Platt scaling can be used). 

# 📌 Applications of SVM
* Text classification  :
Spam detection, sentiment analysis, and categorizing documents.

* Image recognition  :
Handwriting recognition (e.g., digit classification in MNIST), face detection.

* Bioinformatics  :
Protein classification, cancer detection using gene expression data.

* Finance  :
Credit risk analysis, stock market trend prediction.

* Industrial applications  :
Fault detection in machinery, quality control in manufacturing.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x=np.array([
    [2,2],[3,3],[3,4],[5,6],[6,7],[7,8],[1,1],[2,1],[6,5],[7,6]
])
y=np.array([0,0,0,1,1,1,0,0,1,1])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=svm.SVC(kernel='linear')
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("model accuracy:",accuracy)
```
